```python
import turtle
import time #programın yavaşlatılabilmesi için zaman fonksiyonu girildi.
import random #yem yerinin rastgele belirlenmesi için gereklidir.

hiz=0.15 #hız 0.15 olarak belirlendi

pencere=turtle.Screen() #oyunun oynandığı pencere Screen komutuyla oluşturuluyor.
pencere.title("Snake Game") #pencerenin(oyunun) adı title'la oluşturuluyor.
pencere.bgcolor("yellow") #pencerenin rengi belirleniyor. Ben sarı yaptım.
pencere.setup(width=600, height=600) #pencerenin boyutları belirleniyor bu komutla.600 uzunluk, 600 genişlik olacak.
pencere.tracer(0)

    
kafa = turtle.Turtle() #Yılanın kafasını oluşturuyoruz.
kafa.speed(0) #kafanın hızı oyun ilk açıldığında 0 yani hareketsiz olacaktır biz kontrol edeceğiz.
kafa.shape("circle") #kafa daire şeklinde olacaktır.
kafa.color("darkblue") #kafanın rengini belirliyoruz.
kafa.penup() #kafa hareket ederken yazı yazmayacak.
kafa.goto(0, 100) #kafanın durduğu yer belirlenir.
kafa.direction="stop" #kafanın yönü duruyor.

yem = turtle.Turtle() #Yılanın yiyeceği yemi oluşturuyoruz.
yem.speed(0) #yem oyun ilk açıldığında 0 yani hareketsiz olacaktır.
yem.shape("circle") #yem daire şeklinde olacaktır.
yem.color("red") #yemin rengini belirliyoruz.
yem.penup()
yem.goto(0, 0) #yemin durduğu yer belirlenir.
yem.shapesize(0.95, 0.95) #yemin büyüklüğü belirlenir. 

kuyruklar=[] #kuyruk oluşumu için boş liste oluşturulur.

puan = 0

yaz = turtle.Turtle() 
yaz.speed(0) 
yaz.shape("square") 
yaz.color("gray") 
yaz.penup()
yaz.goto(0, 260) 
yaz.hideturtle()  
yaz.write("Puan: ()".format(puan), align="center", font=("Courier", 24, "normal"))

#hareket sisteminin oluşturulması
def move():
    if kafa.direction == "up":
        y=kafa.ycor()
        kafa.sety(y+20)
    if kafa.direction == "down":
        y=kafa.ycor()
        kafa.sety(y-20)
    if kafa.direction == "right":
        x=kafa.xcor()
        kafa.setx(x+20)
    if kafa.direction == "left":
        x=kafa.xcor()
        kafa.setx(x-20)
        
def goUp():
    if kafa.direction !="down":
        kafa.direction="up"
def goDown():
    if kafa.direction !="up":
        kafa.direction="down"
def goRight():
    if kafa.direction !="left":
        kafa.direction="right"        
def goLeft():
    if kafa.direction !="right":
        kafa.direction="left"

#ok tuşlarının belirlenmesi.        
pencere.listen()
pencere.onkey(goUp, "Up")
pencere.onkey(goDown, "Down")
pencere.onkey(goRight, "Right")
pencere.onkey(goLeft, "Left")

while True:
    pencere.update() #pencerenin açık kalmasını sağlar.
    
    if kafa.xcor() > 300 or kafa.xcor() < -300 or kafa.ycor() > 300 or kafa.ycor() < -300:
        time.sleep(1)
        kafa.goto(0, 0)
        kafa.direction = "stop"
        
        for kuyruk in kuyruklar:
            kuyruk.goto(1000, 1000)
            
        kuyruklar = [] #kuyruğun sıfırlanmasını sağlar.
        puan=0
        yaz.clear()
        yaz.write("Puan: ()".format(puan), align="center", font=("Courier", 24, "normal"))
        
        hiz=0.15    
    
    if kafa.distance(yem) < 20: #kafa ile yem arasında 20 pikselden az mesafe kaldıysa eğer yem rastgele olarak yer değiştirir.
        x=random.randint(-250, 250)
        y=random.randint(-250, 250)
        yem.goto(x, y)
        
        puan = puan + 10
        yaz.clear()
        yaz.write("Puan: ()".format(puan), align="center", font=("Courier", 24, "normal"))
        
        hiz = hiz - 0.001
        
        yeniKuyruk=turtle.Turtle() #kuyruk oluşturuluyor.
        yeniKuyruk.speed(0) #oluşturulan kuyruğun ilk baştaki hareketi 0 dır yani yoktur.
        yeniKuyruk.shape("square") #kuyruğun şekli kare olarak eklenecektir.
        yeniKuyruk.color("blue") #kuyruğun rengi mavidir.
        yeniKuyruk.penup()
        kuyruklar.append(yeniKuyruk) #oluşan yeni kuyruk kuyruklar listesine eklenecek.
        
    for i in range(len(kuyruklar) - 1, 0, -1):
        x = kuyruklar[i - 1].xcor()
        y = kuyruklar[i - 1].ycor()
        kuyruklar[i].goto(x, y)
        
    if len(kuyruklar) > 0:
        x = kafa.xcor()
        y = kafa.ycor()
        kuyruklar[0].goto(x, y)
    
    move() #tanımlanan defmove(hareket) fonksiyonunun döngü içinde çalışmasını sağlar. 
    time.sleep(hiz) #programın yavaşlatılması için hız değişkeni atandı.
```
